# **Joining tables around the `order_items` table**

This notebook plays, for fine tuning purpose, with the different forms of merge and reduction of tables.

# Merges without `order_payments` and `order_reviews`

```python
import pandas as pd
from olist_commons import *
def _get_merged_data(how='left'):
    oi = get_order_items()
    oi_o = pd.merge(oi, get_orders(), how=how, on='order_id')
    oi_o_p = pd.merge(oi_o, get_products(), how=how, on='product_id')
    oi_o_p_s = pd.merge(oi_o_p, get_sellers(), how=how, on='seller_id')
    oi_o_p_s_c = pd.merge(oi_o_p_s, get_customers(), how=how, on='customer_id')
    return oi_o_p_s_c
```

## Left merge (using only keys from the root frame `order_items`)

```python
display(_get_merged_data(how='left'))
```

## Right merge (using only keys from the `order_items` children frames)

```python
display(_get_merged_data(how='right'))
```

## Outer merge (using union of keys from both `order_items` and its children frames)

```python
display(_get_merged_data(how='outer'))
```

## Inner merge (using intersection of keys from both `order_items` and its children frames)

```python
display(_get_merged_data(how='inner'))
```

# Complete merge with `order_payments` and `order_reviews`

We take as a basis the outer merge.

```python
import pandas as pd
from olist_commons import *
def _get_merged_data(how='outer'):
    m = get_order_items()
    m = pd.merge(m, get_orders(), how=how, on='order_id')
    m = pd.merge(m, get_products(), how=how, on='product_id')
    m = pd.merge(m, get_sellers(), how=how, on='seller_id')
    m = pd.merge(m, get_customers(), how=how, on='customer_id')
    m = pd.merge(m, get_order_payments(), how=how, on='order_id')    
    return m

display(_get_merged_data())
```
